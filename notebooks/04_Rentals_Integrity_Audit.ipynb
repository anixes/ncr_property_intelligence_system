# 04 - Rentals Data Integrity Audit (Model Stage)

## 📌 Objective
Audit the statistical health of the **Model-Ready** Rentals Dataset (`model_rentals.parquet`).
Here, outliers (Area > 6000, Rent > 5L) are clipped. We verify the cardinal constraints on text data and validate that the `h3_median_price` correctly acts as a Leakage-Proof spatial prior.

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

%matplotlib inline
sns.set_theme(style="whitegrid")

df = pd.read_parquet("../data/model/model_rentals.parquet")
print(f"🔹 Final Rental Model Rows: {len(df):,}")
```

## 1. Categorical Cardinality Check
Reviewing the number of unique Agents and Societies. High cardinality here could lead to overfitting, but XGBoost (especially with depth control) can handle these categorical proxies if encoded properly.

```python
cats = ['city', 'agent_name', 'sector']
print("Unique Category Counts:")
for c in cats:
    if c in df.columns:
        print(f" - {c}: {df[c].nunique():,}")
```

## 2. Engineered Target Consistency Analysis

```python
plt.figure(figsize=(10, 5))
sns.kdeplot(df['price_per_sqft'], fill=True, color='purple', alpha=0.6)
plt.title("Rental Yield Density (Rent per Sq.Ft.)")
plt.xlabel("Rent per Sq.Ft.")
plt.show()
```
