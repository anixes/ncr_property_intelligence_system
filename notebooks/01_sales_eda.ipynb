# NCR Sales EDA & Outlier Strategy

## 📌 Objective
Analyze the model-ready sales dataset (19,147 listings) to validate target distributions and confirm geometric integrity.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from pathlib import Path

%matplotlib inline
sns.set_theme(style="whitegrid")

df = pd.read_parquet("../data/model/model_sales.parquet")
print(f"🔹 Sales Data Columns: {df.columns.tolist()[:10]}")
```

## 1. Price per Sq.Ft. Distribution
We clip the bottom ₹10L and 250 sqft to remove data noise.

```python
sns.histplot(df['price_per_sqft'], bins=50, kde=True)
plt.title("Target Variable: Price per Sq.Ft.")
plt.show()
```

## 2. City-Wise Price Tiers
Gurugram maintains a strong lead in asset valuation.

```python
sns.boxplot(x='city', y='price_per_sqft', data=df)
plt.yscale('log')
plt.xticks(rotation=45)
plt.show()
```
